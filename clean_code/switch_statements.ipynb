```python
from abc import ABC, abstractmethod
class InvalidEmployeeType(Exception):
    def __init__(self, employee_type):
        super().__init__(f"Invalid employee type: {employee_type}")
        self.employee_type = employee_type

class Money(ABC):
    def __init__(self, amount, currency="USD"):
        if amount < 0:
            raise ValueError("Amount cannot be negative")
        self.amount = amount
        self.currency = currency

    @abstractmethod
    def __str__(self):
        """Return a string representation of the money."""
        pass

class Employee(ABC):
    @abstractmethod
    def isPayDay(self):
        pass
    
    @abstractmethod
    def calculatePay(self):
        pass
    
    @abstractmethod
    def deliverPay(self, money):
        pass

class CommissionedEmployee(Employee):
    def __init__(self, employee_record):
        self.record = employee_record
    
    def isPayDay(self):
        return super().isPayDay()
    
    def calculatePay(self):
        return super().calculatePay()
    
    def deliverPay(self, money):
        return super().deliverPay(money)

class HourlyEmployee(Employee):
    def __init__(self, employee_record):
        self.record = employee_record
    
    def isPayDay(self):
        return super().isPayDay()
    
    def calculatePay(self):
        return super().calculatePay()
    
    def deliverPay(self, money):
        return super().deliverPay(money)
    
class SalariedEmployee(Employee):
    def __init__(self, employee_record):
        self.record = employee_record
    
    def isPayDay(self):
        return super().isPayDay()
    
    def calculatePay(self):
        return super().calculatePay()
    
    def deliverPay(self, money):
        return super().deliverPay(money)

class EmployeeFactory(ABC):
    @abstractmethod
    def createEmployee(self, employeeRecord):
        pass

class EmployeeFactoryImp(EmployeeFactory):
    
    def createEmployee(self, employeeRecord):
        if employeeRecord.type == 'Commissioned':
            return CommissionedEmployee
        elif employeeRecord.type == 'Hourly':
            return HourlyEmployee
        elif employeeRecord.type == 'Salaried':
            return SalariedEmployee
        else:
            raise InvalidEmployeeType(employeeRecord.type)
```
